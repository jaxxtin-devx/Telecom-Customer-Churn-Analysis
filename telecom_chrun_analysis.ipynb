```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
print("✅libraries installed usccessfully👌")
```

```python
from google.colab import files
import pandas as pd

# Upload file
uploaded = files.upload()

df=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
print("shape of the dataset",df.shape)
print("\nColumn Names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
display(df.head())
print("\n data types",df.dtypes)
print("\n missing valuse", df.isnull().sum())
```

```python
churn_rate = df['Churn'].value_counts()
churn_percent = df['Churn'].value_counts(normalize=True)*100

print("Churn rate",churn_rate)
print("\n Churn percent")
print(round(churn_percent,2))
```

```python
# Visualization 1: Churn Distribution
plt.figure(figsize=(6,5))
sns.countplot(data=df, x='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Overall Customer Churn Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Summary Statistics for Numerical Columns
print("\nNumerical Features Summary:")
print(df[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

for col in num_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Histogram
    sns.histplot(df[col], kde=True, bins=30, color='#3498db', ax=axes[0])
    axes[0].set_title(f'Distribution of {col}', fontsize=13, fontweight='bold')

    # Boxplot
    sns.boxplot(y=df[col], color='#2ecc71', ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}', fontsize=13, fontweight='bold')

    plt.tight_layout()
    plt.show()

    # Summary stats
    print(f"→ {col} | Min: {df[col].min():.2f} | Max: {df[col].max():.2f} | Mean: {df[col].mean():.2f}\n")
```

```python
# Fix Churn column data type
print("Before fix - Churn dtype:", df['Churn'].dtype)
print("Unique values in Churn:", df['Churn'].unique())

# Re-convert Churn to numeric (0 and 1)
df['Churn'] = df['Churn'].map({'No': 0, 'Yes': 1})

print("\nAfter fix - Churn dtype:", df['Churn'].dtype)
print("Unique values now:", df['Churn'].unique())
```

```python
# Important Categorical Columns
cat_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService',
            'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup',
            'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies',
            'Contract', 'PaperlessBilling', 'PaymentMethod']

# Create count plots with Churn hue
for col in cat_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x=col, hue='Churn', palette=['#66b3ff', '#ff9999'])
    plt.title(f'{col} vs Churn', fontsize=14, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45)
    plt.legend(title='Churn', labels=['No', 'Yes'])
    plt.show()

    # Print Churn Rate safely
    churn_by_cat = df.groupby(col)['Churn'].mean() * 100
    print(f"\n📊 Churn Rate (%) by {col}:")
    print(round(churn_by_cat, 2))
    print("-" * 60)
```

```python
# ====================== 1. Churn by Contract Type ======================
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Contract Type', fontsize=14, fontweight='bold')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.show()

print("Churn Rate by Contract Type:")
print(round(df.groupby('Contract')['Churn'].mean()*100, 2))
```

```python
# ====================== 2. Churn by Internet Service ======================
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='InternetService', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Internet Service', fontsize=14, fontweight='bold')
plt.show()

print("\nChurn Rate by Internet Service:")
print(round(df.groupby('InternetService')['Churn'].mean()*100, 2))

# ====================== 3. Churn by Payment Method ======================
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='PaymentMethod', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Payment Method', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.show()

print("\nChurn Rate by Payment Method:")
print(round(df.groupby('PaymentMethod')['Churn'].mean()*100, 2))

# ====================== 4. Tenure vs Churn ======================
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', palette=['#66b3ff', '#ff9999'], bins=30)
plt.title('Tenure (Months) Distribution - Churn vs Non-Churn', fontsize=14, fontweight='bold')
plt.xlabel('Tenure (Months)')
plt.ylabel('Number of Customers')
plt.show()

# ====================== 5. Monthly Charges vs Churn ======================
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette=['#66b3ff', '#ff9999'])
plt.title('Monthly Charges Distribution - Churn vs Non-Churn', fontsize=14, fontweight='bold')
plt.show()

print("\nAverage Monthly Charges:")
print(round(df.groupby('Churn')['MonthlyCharges'].mean(), 2))
```

```python
# Final Cleaning for TotalCharges
print("Before fix - TotalCharges dtype:", df['TotalCharges'].dtype)

# Convert again forcefully
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill any remaining missing values
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

print("After fix - TotalCharges dtype:", df['TotalCharges'].dtype)
print("Any missing values left?", df['TotalCharges'].isnull().sum())
```

```python
# ====================== FEATURE ENGINEERING ======================
df['tenure_group'] = pd.cut(df['tenure'],
                           bins=[0, 12, 24, 48, 100],
                           labels=['0-12 months', '13-24 months', '25-48 months', '49+ months'])

# ====================== VISUALIZATIONS ======================

# 1. Churn Rate by Tenure Group
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='tenure_group', y='Churn', palette='Reds_r')
plt.title('Churn Rate by Tenure Group', fontsize=14, fontweight='bold')
plt.xlabel('Tenure Group')
plt.ylabel('Churn Rate (%)')
plt.show()

print("Churn Rate by Tenure Group (%):")
print(round(df.groupby('tenure_group')['Churn'].mean()*100, 2))

# 2. Churn Rate by Senior Citizen
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='SeniorCitizen', y='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate: Senior Citizens vs Non-Senior', fontsize=14, fontweight='bold')
plt.ylabel('Churn Rate (%)')
plt.show()

# 3. Correlation Heatmap (Now Safe)
plt.figure(figsize=(8,6))
corr = df[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Numerical Features', fontsize=14, fontweight='bold')
plt.show()

# ====================== KEY INSIGHTS SUMMARY ======================
print("\n" + "="*65)
print("🔥 KEY BUSINESS INSIGHTS")
print("="*65)
print(f"Overall Churn Rate                  : {round(df['Churn'].mean()*100, 2)}%")
print(f"Month-to-month Contract Churn       : {round(df[df['Contract']=='Month-to-month']['Churn'].mean()*100, 2)}%")
print(f"Fiber Optic Internet Churn          : {round(df[df['InternetService']=='Fiber optic']['Churn'].mean()*100, 2)}%")
print(f"Electronic Check Payment Churn      : {round(df[df['PaymentMethod']=='Electronic check']['Churn'].mean()*100, 2)}%")
print(f"New Customers (0-12 months) Churn   : {round(df[df['tenure_group']=='0-12 months']['Churn'].mean()*100, 2)}%")
```

```python
# Set style for all plots
sns.set_style("whitegrid")

# ====================== FINAL PROFESSIONAL VISUALIZATIONS ======================

# 1. Churn by Contract Type (Most Important)
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df, x='Contract', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Contract Type (Highest Insight)', fontsize=15, fontweight='bold')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.legend(title='Churn', labels=['No', 'Yes'])

# Add percentage on bars
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2., height + 3, f'{height:.0f}', ha="center")
plt.show()

# 2. Churn by Internet Service
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='InternetService', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Internet Service Type', fontsize=15, fontweight='bold')
plt.show()

# 3. Payment Method
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='PaymentMethod', hue='Churn', palette=['#66b3ff', '#ff9999'])
plt.title('Churn Rate by Payment Method', fontsize=15, fontweight='bold')
plt.xticks(rotation=45)
plt.show()

# 4. Services that Reduce Churn (Very Good Insight)
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport']
plt.figure(figsize=(12, 6))
for i, service in enumerate(services, 1):
    plt.subplot(2, 2, i)
    sns.barplot(data=df, x=service, y='Churn', palette='Blues_r')
    plt.title(f'Churn Rate with {service}')
    plt.ylabel('Churn Rate')
plt.tight_layout()
plt.show()

print("✅ All Final Visualizations Created Successfully!")
```
